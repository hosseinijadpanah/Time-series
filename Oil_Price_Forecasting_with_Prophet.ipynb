```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Step 1: Read the data and preprocess
df = pd.read_csv("oil price updated dataset.csv")
df["Date"] = pd.to_datetime(df["Date"])

# Step 2: Prepare data for Prophet
data = df[["Date", "Open"]].rename(columns={"Date": "ds", "Open": "y"})

# Step 3: Create and train the Prophet model
model = Prophet()
model.fit(data)

# Step 4: Make future predictions
future = model.make_future_dataframe(periods=30)  # Adjust the number of periods as needed
forecast = model.predict(future)

# Step 5: Plot the results
fig = model.plot(forecast)
plt.xlabel("Date")
plt.ylabel("Open Price")
plt.title("Prophet Forecasting")
plt.show()
```
